Data analysis:

Youtube Categories

```python
import matplotlib.pyplot as plt
import numpy as np


# 'Category' is the column name in your DataFrame
category_counts = data['Category'].value_counts()

# Define colors for the bars
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'black', 'pink'] 

# Plotting
plt.bar(category_counts.index, category_counts, color=colors)
plt.xlabel('Category')
plt.ylabel('Counts')
plt.title('Types of Youtube categories')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.show()
```

After analzying the bar graph, it is clear that the most popular Youtube categories are music and entertainment indicating that viewers are interested in that type of content and therefore the youtube alogrithm recommends those users that type of content (in their youtube recommended). 

Youtube channels such as T-series and MrBeast fall under these categories since they are the most popular Youtube channels with a high following.

However, the least popular Youtube categories are news and games which indicates that Youtube users are less interested in that type of content, since they prefer to watch music and entertainment content, therefore the youtube alogrithm recommends that type of content less to users. Therefore, these types of channels become less popular and are less in quantity. 

Types of languages

```python
# Assuming 'Primary language' is the column name in your DataFrame
primary_language_counts = data['Primary language'].value_counts()

# Plotting
plt.figure(figsize=(18, 20))  # Define the size of the pie chart
plt.pie(primary_language_counts.values, labels=primary_language_counts.index, autopct='%1.1f%%', startangle=140)

# Title
plt.title('Distribution of Primary Languages')

# Show plot
plt.show()
```

It is clear that the most spoken languages to make Youtube videos is English, Hindi, and Spanish. However, the least spoken languages to make Youtube videos are Urdu, Portuguese, and Bhojpuri.

Distribution of countries

```python
# 'Country' is the column name in your DataFrame
country_counts = data['Country'].value_counts()

# Define colors for the bars
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'black', 'pink', 'indigo', 'violet', 'yellow'] 

# Plotting
plt.bar(country_counts.index, country_counts, color=colors)
plt.xlabel('Countries')
plt.ylabel('Quantity')
plt.title('Distribution of countries')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Show plot
plt.show()
```

By anaylysing the distribution of countries, India has the highest number of Youtube channels 

Number of brand channels


```python
# 'Brand channel' is the column name in your DataFrame
brandchannel_counts = data['Brand channel'].value_counts()

# Define colors for the pie chart
colors = ['blue', 'red']

# Plotting
plt.pie(brandchannel_counts, labels=brandchannel_counts.index, colors=colors, autopct='%1.1f%%')
plt.title('Distribution of Brand Channels')

# Show plot
plt.show()
```

Based on the analysis of this bar graph, it is evident that approximately 64% YouTube channels are branded, while the number of non-branded channels falls within the range of 36%. This indicates that a significant portion of the YouTube channels in this dataset are associated with a specific brand.
